```python
import csv
import pandas as pd
df = pd.read_csv('trips_data_linxiao.csv', encoding = "ISO-8859-1")
```

```python
import seaborn as sns
```

```python
import datetime

hour = []
ALL_day = []
date = list(df["Request Time"])
for i in range(len(date)):
    time = date[i][:10]
    h = date[i][10:13]
    day = datetime.datetime.strptime(time, '%Y-%m-%d').strftime('%A')
    hour.append(int(h))
    ALL_day.append(day)
print(hour)
```

```python
day_fre = {}
day_fre['Monday'] = 0
day_fre['Tuesday']= 0
day_fre['Wednesday'] = 0
day_fre['Thursday'] = 0
day_fre['Friday'] = 0
day_fre['Saturday'] = 0
day_fre['Sunday'] = 0
for i in ALL_day:
    if i not in day_fre:
        day_fre[i] =1
    else:
        day_fre[i] +=1
day = ['Mon','Tue', 'Wed', 'Thu', 'Fri','Sat','Sun']
day_fre
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = day

performance = [day_fre['Monday'],day_fre['Tuesday'],day_fre['Wednesday'],day_fre['Thursday'],day_fre['Friday'],day_fre['Saturday'],day_fre['Sunday']]
y_pos = np.arange(len(day))
barlist1 = plt.bar(y_pos[:5], performance[:5], align='center', alpha=0.5, color = "blue" ,label = "Weekday")
barlist2 = plt.bar(y_pos[5:7],performance[5:7],align='center', alpha=0.5, color = "red" ,label = "Weekends")
plt.xticks(y_pos, objects)
plt.ylabel('Times')
plt.title('Frequecy using Uber by day')

plt.legend()

plt.show()
```

```python
df['Fare Amount']
day_cost = {}
day_cost['Monday'] = 0
day_cost['Tuesday']= 0
day_cost['Wednesday'] = 0
day_cost['Thursday'] = 0
day_cost['Friday'] = 0
day_cost['Saturday'] = 0
day_cost['Sunday'] = 0
for i in range(len(ALL_day)):
    a = float( df['Fare Amount'][i])
    day_cost[ALL_day[i]] += a
day_cost    
```

```python
objects = day

performance = [day_cost['Monday'],day_cost['Tuesday'],day_cost['Wednesday'],day_cost['Thursday'],day_cost['Friday'],day_cost['Saturday'],day_cost['Sunday']]
y_pos = np.arange(len(day))

barlist1 = plt.bar(y_pos[:5], performance[:5], align='center', alpha=0.5, color = "blue" ,label = "Weekday")
barlist2 = plt.bar(y_pos[5:7],performance[5:7],align='center', alpha=0.5, color = "red" ,label = "Weekends")
plt.xticks(y_pos, objects)
plt.ylabel('Cost(USD)')
plt.title('Cost by day of week')
plt.legend()
plt.show()
```
